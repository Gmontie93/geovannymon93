```python
import pandas as pd

df = pd.read_csv('./medical_raw_data.csv', index_col=0)
df.info()
for col in df.columns:
    print(f"--------{col}----------")
    # Check the data type of each column individually and the number of entries.
    print('Check the data type of each column individually')
    print(df[col].info())

    #Check the entries of the columns.
    print('--Check the entries of the columns--')
    print(df[col].unique())

    #check the quantity of each value. 
    print('--check the quantity of each value--')
    print(df[col].value_counts())
    
    #Check the number of missing entries.
    print('--Check the number of missing entries--')
    print(f"Number of missing values in {col}:", df[col].isnull().sum())

    #Check if data is within reasonable bounds.
    print('--Check if data is within reasonable bounds--')
    print(df[col].describe())
    
    




```
